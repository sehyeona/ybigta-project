종류 
* standardscaler : 정규분포 이용
* minmaxscaler : 최대/최소값이 각각 0, 1
* maxabsscaler : 최대절대값과 0이 각각 1, 0 이 되도록하는 scaling
* robustscaler : median과 IQR 사용 outlier의 영향을 최소화 한다 

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
import sys
print(sys.executable)
```

```python
data = pd.read_csv("./data/train.csv")
test_data = pd.read_csv("./data/test.csv")
data.head()
```

```python
# 스케일링 변수에서 id type fiberid 제거

data_for_scaling = data[data.columns[3:]]
data_for_scaling.head()
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
# type과 id fiberid를 제외하고 학습
std_scaler.fit(data_for_scaling)
# 학습후 변환
train_stdscaled = std_scaler.transform(data_for_scaling)
train_stdscaled
# 학습후 변환한 데이터를 다시 원래 데이터로 만들기
data_stdscaled = pd.DataFrame(train_stdscaled, columns = data_for_scaling.columns)
data_stdscaled = pd.concat([data[['id','type', 'fiberID']], data_stdscaled], axis=1)
data_stdscaled.head()
```

```python
# 테스트 변형
test_stdscaled = std_scaler.transform(test_data[test_data.columns[2:]])
testdata_stdscaled = pd.DataFrame(test_stdscaled, columns = data_for_scaling.columns)
testdata_stdscaled = pd.concat([test_data[['id','fiberID']], testdata_stdscaled], axis=1)
testdata_stdscaled.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
mm_scaler = MinMaxScaler()
# type과 id를 제외하고 학습
mm_scaler.fit(data_for_scaling)
# 학습후 변환
train_mmscaler = mm_scaler.transform(data_for_scaling)
train_mmscaler
# 학습후 변환한 데이터를 다시 원래 데이터로 만들기
data_mmscaled = pd.DataFrame(train_stdscaled, columns = data_for_scaling.columns)
data_mmscaled = pd.concat([data[['id','type', 'fiberID']], data_mmscaled], axis=1)
data_mmscaled.head()
```

```python
# 테스트 변형
test_mmscaled = std_scaler.transform(test_data[test_data.columns[2:]])
testdata_mmscaled = pd.DataFrame(test_mmscaled, columns = data_for_scaling.columns)
testdata_mmscaled = pd.concat([test_data[['id','fiberID']], testdata_mmscaled], axis=1)
testdata_mmscaled.head()
```

```python
from sklearn.preprocessing import MaxAbsScaler
ma_scaler = MaxAbsScaler()
# type과 id를 제외하고 학습
ma_scaler.fit(data_for_scaling)
# 학습후 변환
train_mascaled = ma_scaler.transform(data_for_scaling)
train_mascaled
# 학습후 변환한 데이터를 다시 원래 데이터로 만들기
data_mascaled = pd.DataFrame(train_mascaled, columns = data_for_scaling.columns)
data_mascaled = pd.concat([data[['id','type', 'fiberID']], data_mascaled], axis=1)
data_mascaled.head()
```

```python
# 테스트 변형
test_mascaled = std_scaler.transform(test_data[test_data.columns[2:]])
testdata_mascaled = pd.DataFrame(test_mascaled, columns = data_for_scaling.columns)
testdata_mascaled = pd.concat([test_data[['id','fiberID']], testdata_mascaled], axis=1)
testdata_mascaled.head()
```

```python
from sklearn.preprocessing import RobustScaler
rb_scaler = RobustScaler()
# type과 id를 제외하고 학습
rb_scaler.fit(data_for_scaling)
# 학습후 변환
train_rbscaled = rb_scaler.transform(data_for_scaling)
train_rbscaled
# 학습후 변환한 데이터를 다시 원래 데이터로 만들기
data_rbscaled = pd.DataFrame(train_rbscaled, columns = data_for_scaling.columns)
data_rbscaled = pd.concat([data[['id','type', 'fiberID']], data_rbscaled], axis=1)
data_rbscaled.head()

```

```python
# 테스트 변형
test_rbscaled = std_scaler.transform(test_data[test_data.columns[2:]])
testdata_rbscaled = pd.DataFrame(test_rbscaled, columns = data_for_scaling.columns)
testdata_rbscaled = pd.concat([test_data[['id','fiberID']], testdata_rbscaled], axis=1)
testdata_rbscaled.head()
```

```python
from lightgbm import LGBMClassifier
import pickle

# max min 에 사용할 모델
lgbm_model_mm = LGBMClassifier(boosting_type='gbdt', objective='binary', num_leaves=10,
                                learning_rate=0.1, n_estimators=2000, max_depth=15,
                                bagging_fraction=0.9, feature_fraction=0.9, reg_lambda=0.2)

# max abs 에 사용할 모델
lgbm_model_ma = LGBMClassifier(boosting_type='gbdt', objective='binary', num_leaves=10,
                                learning_rate=0.1, n_estimators=2000, max_depth=15,
                                bagging_fraction=0.9, feature_fraction=0.9, reg_lambda=0.2)

# robust 에 사용할 모델
lgbm_model_robust = LGBMClassifier(boosting_type='gbdt', objective='binary', num_leaves=10,
                                learning_rate=0.1, n_estimators=2000, max_depth=15,
                                bagging_fraction=0.9, feature_fraction=0.9, reg_lambda=0.2)
```

```python
X_mm = data_mmscaled[data.columns[2:]]
y_mm = data_mmscaled['type']
```

```python
lgbm_model_mm.fit(X_mm, y_mm)
```

```python
pickle.dump(lgbm_model_mm, open('./models/lgbm_model_mm.sav', 'wb'))
```

```python
X_ma = data_mascaled[data.columns[2:]]
y_ma = data_mascaled['type']
```

```python
lgbm_model_ma.fit(X_ma, y_ma)
```

```python
pickle.dump(lgbm_model_ma, open('./models/lgbm_model_ma.sav', 'wb'))
```

```python
X_rb = data_rbscaled[data.columns[2:]]
y_rb = data_rbscaled['type']
```

```python
lgbm_model_robust.fit(X_rb, y_rb)
```

```python
pickle.dump(lgbm_model_robust, open('./models/lgbm_model_rb.sav', 'wb'))
```

```python
id_for_index = test_data['id']
test_mm = testdata_mmscaled[test_data.columns[1:]]
test_ma = testdata_mascaled[test_data.columns[1:]]
test_rb = testdata_rbscaled[test_data.columns[1:]]

sample = pd.read_csv("./data/sample_submission.csv")
col_order = sample.columns[1:]
```

```python
predictions_mm = lgbm_model_mm.predict_proba(X = test_mm)
result_mm = pd.DataFrame(data=predictions_mm, index=id_for_index, columns=lgbm_model_mm.classes_)
result_mm = result_mm[col_order]
result_mm.to_csv('./result/lgbm_mmscaler.csv', sep=',')
```

```python
predictions_ma = lgbm_model_ma.predict_proba(X = test_ma)
result_ma = pd.DataFrame(data=predictions_ma, index=id_for_index, columns=lgbm_model_ma.classes_)
result_ma = result_ma[col_order]
result_ma.to_csv('./result/lgbm_mascaler.csv', sep=',')
```

```python
predictions_rb = lgbm_model_robust.predict_proba(X = test_rb)
result_rb = pd.DataFrame(data=predictions_rb, index=id_for_index, columns=lgbm_model_rb.classes_)
result_rb = result_rb[col_order]
result_rb.to_csv('./result/lgbm_rbscaler.csv', sep=',')
```
