모든 스케일러는 sklearn.preprocessing 안에 각자 이름으로 들어있음

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
import sys
print(sys.executable)
```

```python
data = pd.read_csv("./data/train.csv")
test_data = pd.read_csv("./data/test.csv")
data.head()
```

스케이링 하는 변수에 id 와 type은 제거해주었다. (일단 fiberid는 보류)

```python
data_for_scaling = data[data.columns[2:]]
data_for_scaling.head()
```

* Standard Scaler 사용

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
# type과 id를 제외하고 학습
std_scaler.fit(data_for_scaling)
# 학습후 변환
train_stdscaler = std_scaler.transform(data_for_scaling)
train_stdscaler
# 학습후 변환한 데이터를 다시 원래 데이터로 만들기
data_stdscaler = pd.DataFrame(train_stdscaler, columns = data_for_scaling.columns)
data_stdscaler = pd.concat([data[['id','type']], data_stdscaler], axis=1)
data_stdscaler.head()
```

* Min Max scaler 사용

```python
from sklearn.preprocessing import MinMaxScaler
mm_scaler = MinMaxScaler()
# type과 id를 제외하고 학습
mm_scaler.fit(data_for_scaling)
# 학습후 변환
train_mmscaler = mm_scaler.transform(data_for_scaling)
train_mmscaler
# 학습후 변환한 데이터를 다시 원래 데이터로 만들기
data_mmscaler = pd.DataFrame(train_stdscaler, columns = data_for_scaling.columns)
data_mmscaler = pd.concat([data[['id','type']], data_mmscaler], axis=1)
data_mmscaler.head()
```

* Max Abs Scaler 사용

```python
from sklearn.preprocessing import MaxAbsScaler
ma_scaler = MaxAbsScaler()
# type과 id를 제외하고 학습
ma_scaler.fit(data_for_scaling)
# 학습후 변환
train_mascaler = ma_scaler.transform(data_for_scaling)
train_mascaler
# 학습후 변환한 데이터를 다시 원래 데이터로 만들기
data_mascaler = pd.DataFrame(train_mascaler, columns = data_for_scaling.columns)
data_mascaler = pd.concat([data[['id','type']], data_mascaler], axis=1)
data_mascaler.head()
```

```python
import catboost as cb
cb_model_std = cb.CatBoostClassifier()
cb_model_mm = cb.CatBoostClassifier()
cb_model_ma = cb.CatBoostClassifier()
```

```python
X_std = data_stdscaler[data.columns[2:]]
X_mm = data_mmscaler[data.columns[2:]]
X_ma = data_mascaler[data.columns[2:]]
y = data['type']
```

```python
cb_model_std.fit(X_std, y)
```

```python
cb_model_mm.fit(X_mm, y)
```

```python
cb_model_ma.fit(X_ma, y)
```

```python
from sklearn.externals import joblib 
# 객체를 pickled binary file 형태로 저장한다 
# std mm ma로 scaling 하여 학습시킨 모델들을 각각 sacling 의 이름으로 저장하였다.
joblib.dump(cb_model_std, './models/cb_model_basic_std.pkl')
joblib.dump(cb_model_mm, './models/cb_model_basic_mm.pkl') 
joblib.dump(cb_model_ma, './models/cb_model_basic_ma.pkl')
```

```python
import pickle
# save the model to disk
filename = 'finalized_model.sav'
pickle.dump(cb_model_std, open('./models/cb_model_basic_std.sav', 'wb'))
pickle.dump(cb_model_mm, open('./models/cb_model_basic_mm.sav', 'wb'))
pickle.dump(cb_model_ma, open('./models/cb_model_basic_ma.sav', 'wb'))
```

```python
data_stdscaler.to_csv('./scaleddata/std_scaled_with_fiberid.csv', sep=',', index = False)
data_mmscaler.to_csv('./scaleddata/mm_scaled_with_fiberid.csv', sep=',', index = False)
data_mascaler.to_csv('./scaleddata/ma_scaled_with_fiberid.csv', sep=',', index = False)
```
