```python
import pandas as pd
import numpy as np
import sys
print(sys.executable)
```

```python
data = pd.read_csv('./data/train.csv')
X = data.iloc[:,2:]
y = data['type']
X.head()
```

```python
from sklearn.model_selection import train_test_split
from lightgbm import LGBMClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import log_loss, make_scorer

# 모델 기본 틀 생성
lgbm_model = LGBMClassifier(
    objective='binary',bagging_fraction=0.9, num_leaves=10, feature_fraction=0.9, 
     bagging_freq = 1, reg_lambda=0.2, silent=False)

# 그리드 서치 인자 조합 생성
param_dist = { "boosting_type": ['rf', 'gbdt', 'dart', 'goss'],
              "max_depth" : [15, 30, 45],
              "learning_rate" : [0.01 , 0.05, 0.1],
              "n_estimators" : [1500, 2000, 2500],
            }

# 점수를 위한 로그로스 객체 생성
LogLoss = make_scorer(log_loss, greater_is_better=False, needs_proba=True)

# lgbm_model = LGBMClassifier(boosting_type='gbdt', objective='binary', num_leaves=10,
#                                 learning_rate=0.1, n_estimators=2000, max_depth=15,
#                                 bagging_fraction=0.9, feature_fraction=0.9, reg_lambda=0.2)
```

```python
grid_search = GridSearchCV(lgbm_model, n_jobs=-1, param_grid = param_dist, cv=3, scoring=LogLoss)
```

```python
grid_search.fit(X, y)
```

```python
grid_search.best_estimator_
grid_search.best_score_
```
